## You will need to create an API key

For google gemini you will have to setup a google gemini key https://ai.google.dev/gemini-api/docs/quickstart

This is straightforward and free. You will need to make sure you add it to all areas in the below code detailed **ENTER your API KEY**


```python
from google import genai

# The client gets the API key from the environment variable `GEMINI_API_KEY`.
client = genai.Client(api_key="Enter your API Key")

response = client.models.generate_content(
    model="gemini-3-flash-preview", contents="create me a mayo recipe"
)
print(response.text)
```

```python
from google.genai import types
import pathlib
import httpx

client = genai.Client(api_key="Enter your API Key")

# Retrieve and encode the PDF byte
file_path = pathlib.Path('Downloads/sample-local-pdf.pdf')

# Upload the PDF using the File API
sample_file = client.files.upload(
    file=file_path,
)

prompt="Summarize this document"

response = client.models.generate_content(
    model="gemini-3-flash-preview",
    contents=[sample_file, "Summarize this document"])
print()
```

```python
with open("demofile.txt", "a") as f:
  f.write(response.text)
```
